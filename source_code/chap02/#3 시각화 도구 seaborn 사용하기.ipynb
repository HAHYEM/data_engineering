```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import seaborn as sns
```

```python
x = np.linspace(0, 14, 100)

y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)

sns.set_style('whitegrid') # seaborn의 whitegrid 스타일 지정하기

plt.figure(figsize = (10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
sns.set_style('whitegrid')

tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize = (8, 6))
sns.boxplot(x='day', y='total_bill', data=tips) #seaborn
# 박스 그래프 로 보기. x축은 day(요일), y축은 전체 금액(total_bill)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='Set3') # 흡연자(smoker)로 추가 구분하기
plt.show()
```

```python
sns.set_style('darkgrid') # darkgrid 스타일의 그래프 그리기
sns.lmplot(x='total_bill', y='tip', data=tips, size=7) # x축 결제금액, y축 종업원에게 주는 팁
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, palette='Set1', size=7)
plt.show()
```
