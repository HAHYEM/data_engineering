```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import seaborn as sns
```

```python
x = np.linspace(0, 14, 100)

y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)

sns.set_style('whitegrid') # seaborn whitegrid  

plt.figure(figsize = (10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
sns.set_style('whitegrid')

tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize = (8, 6))
sns.boxplot(x='day', y='total_bill', data=tips) #seaborn
#    . x day(), y  (total_bill)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='Set3') # (smoker)  
plt.show()
```

```python
sns.set_style('darkgrid') # darkgrid   
sns.lmplot(x='total_bill', y='tip', data=tips, size=7) # x , y   
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, palette='Set1', size=7)
plt.show()
```
